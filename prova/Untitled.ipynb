```python
mu, sigma = 100, .1
```

```python
h = np.random.normal(mu, sigma, 10000)
len(h)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
count, bins, patches  = plt.hist(h, 100, density = True)
```

```python
count, bins, patches  = plt.hist(h, 100, density = True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2)), color="r")
plt.show()
```
