```python
from pandas import *
import pandas as pd
```

```python
data = pd.read_csv('PluviometriaPaperopoli_4.txt')
data
```

```python
data.columns
```

```python
import datetime as datetime
#funzione che trasforma il vettore anno in una data
map(datetime, data['anno'])
#map applica la funzione (a primo membro) al vettore (a secondo membro) e a tutti i sottomembri
```

```python
data = data.set_index('anno')
data
```

```python
data['1h'][1979]
```

```python
import matplotlib.pyplot as plt
#(%) magic command di jupyter. Mette in linea i comandi
%matplotlib inline
```

```python
# ax axes (tavola da disegno)
ax = data.plot(title="Piogge a Paperopoli", style="o-")
ax.set_ylabel("Precipitazione [mm]")
```

```python
ax = data.plot(kind="bar")
ax.set_ylabel("Precipitazione [mm]")
```
