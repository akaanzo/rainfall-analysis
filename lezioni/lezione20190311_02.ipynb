```python
import numpy as np
```

```python
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 10000)
#s è un vettore che contiene i dati con media 0 e varianza 0.1, che verranno restituiti secondo curva normale
```

```python
np.mean(s)
```

```python
abs(mu - np.mean(s)) < .01
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Plotto un istogramma con i dati di s raggruppati in 30 scatole (bins) contenenti un numero di elementi(ignored) normalizzato (integrale pari a 1) e la curva di Gauss

```python
count, bins, ignored = plt.hist(s, 30, normed = True)
#plt.plot(bins)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.xlabel('x')
plt.ylabel('pdf(x)')
plt.show()
```

```python
x = np.linspace(bins.min(), bins.max(), num = 100) 
```

```python
#import numpy as np
import scipy.stats as stats
import pylab as pl

#ordino i dati dal minimo al massimo
h = sorted([186, 176, 158, 180, 186, 168, 168, 164, 178, 170, 189, 195, 172,
     187, 180, 186, 185, 168, 179, 178, 183, 179, 170, 175, 186, 159,
     161, 178, 175, 185, 175, 162, 173, 172, 177, 175, 172, 177, 180])  #sorted

#APPLICO IL METODO DEI MOMENTI

#calcolo media e dev std di h
fet = stats.norm.pdf(h, np.mean(h), np.std(h))  #this is a fitting indeed

#disegno la curva
pl.plot(h,fet,'-o')

pl.hist(h,density=True)      #use this to draw histogram of your data

plt.xlabel('x')
plt.ylabel('pdf(x)')

pl.show()  
```
