```python
import numpy as np
```

```python
a, b = 0, 1
```

```python
s = np.random.gumbel(a, b, 1000)
#s
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def myGumbel(x,a,b):
    return np.exp(-(x-a)/b)*np.exp(-np.exp(-(x-a)/b))*(1/b)
```

```python
count, bins, patches = plt.hist(s, 50, normed="density")
x = np.linspace(bins.min(), bins.max(), num=100)
plt.plot(x, myGumbel(x,a,b), linewidth=3, color='g')
plt.xlabel("x")
plt.ylabel("p(x)")
plt.title("Distribuzione di Gumbel")
plt.show()
#***********************************************************
count, bins, patches = plt.hist(s, 50, normed="density")
plt.plot(bins, myGumbel(bins,a,b), linewidth=3, color='r')
plt.xlabel("x")
plt.ylabel("p(x)")
plt.title("Distribuzione di Gumbel con bins")
plt.show()
```
